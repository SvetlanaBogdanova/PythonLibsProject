## Задание 1

```python
import numpy as np
from matplotlib import pyplot as plt

%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
x = [1, 2, 3, 4, 5, 6, 7]

y = [3.5, 3.8, 4.2, 4.5, 5, 5.5, 7]
```

```python
plt.plot(x, y)

label_font = {
    "fontsize": 10, 
    "family": "serif",
}
plt.xlabel("x", fontdict=label_font)
plt.ylabel("y", fontdict=label_font)

plt.axis([1, 7, 3.5, 7])

plt.grid(color="lightgrey")

plt.show()
```

```python
plt.scatter(x, y)
plt.show()
```

## Задание 2

```python
t = np.linspace(0, 10, 51)
t
```

```python
f = np.cos(t)
```

```python
label_font = {
    "fontsize": 10, 
    "family": "serif"
}

plt.plot(t, f, label="f(t)", color="green")

plt.title("График f(t)", 
          fontsize=14, 
          fontweight="bold", 
          color="#808080", 
          family="serif")
plt.xlabel("Значения t", fontdict=label_font)
plt.ylabel("Значения f", fontdict=label_font)

plt.axis([0.5, 9.5, -2.5, 2.5])

plt.grid(color="lightgrey")

plt.show()
```

## Задание 3

```python
x = np.linspace(-3, 3, 51)
x
```

```python
y1 = x ** 2
```

```python
y2 = 2 * x + 0.5
```

```python
y3 = -3 * x - 1.5
```

```python
y4 = np.sin(x)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)
ax1, ax2, ax3, ax4 = ax.flatten()

fig.set_size_inches(8, 6)
fig.subplots_adjust(wspace=0.3, hspace=0.3)

ax1.plot(x, y1)
ax1.set_xlim([-5, 5])
ax1.set_title("График y1", fontsize=10, fontweight="bold")

ax2.plot(x, y2)
ax2.set_title("График y2", fontsize=10, fontweight="bold")

ax3.plot(x, y3)
ax3.set_title("График y3", fontsize=10, fontweight="bold")

ax4.plot(x, y4)
ax4.set_title("График y4", fontsize=10, fontweight="bold")
```
